```python
import pandas as pd

# Muat file Excel untuk memeriksa struktur dan isinya
file_path = '/content/Data BPS.xlsx'
data = pd.read_excel(file_path)

# Tampilkan beberapa baris pertama dan informasi ringkasan dari dataset
data_head = data.head()
data_info = data.info()

data_head, data_info
```

```python
# Hapus kolom dengan nilai NaN sebagian besar atau baris header yang tidak relevan
data_cleaned = data.dropna(how='all', axis=1)  # Drop completely empty columns
data_cleaned = data_cleaned.drop(index=[0, 1, 2])  # Drop the first few rows with irrelevant headers

# Setel ulang indeks
data_cleaned.reset_index(drop=True, inplace=True)

# Tampilkan struktur dataset yang telah dibersihkan dan beberapa baris pertama untuk inspeksi lebih lanjut
data_cleaned_head = data_cleaned.head()
data_cleaned_info = data_cleaned.info()

data_cleaned_head, data_cleaned_info
```

```python
# Hapus kolom yang hampir seluruhnya kosong
threshold = len(data_cleaned) * 0.5
data_reduced = data_cleaned.dropna(thresh=threshold, axis=1)

# Tampilkan struktur setelah pembersihan lebih lanjut dan beberapa baris pertama
data_reduced_info = data_reduced.info()
data_reduced_head = data_reduced.head()

data_reduced_info, data_reduced_head
```

```python

# Memeriksa tipe data dan nilai untuk mengidentifikasi kolom dengan data numerik dan kategorikal

# Mengonversi kolom ke tipe numerik jika memungkinkan, dengan memaksa kesalahan untuk mengidentifikasi masalah
data_transformed = data_reduced.apply(pd.to_numeric, errors='coerce')

# Membandingkan dengan data asli untuk mengidentifikasi kolom yang sepenuhnya numerik dan kolom yang memerlukan pengkodean
numeric_cols = data_transformed.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = data_reduced.select_dtypes(include=['object']).columns.tolist()

numeric_cols, categorical_cols
```

```python
# konversi numerik pada kolom yang sesuai di `data_reduced`, sambil mempertahankan nilai asli di `data_reduced`

# Konversi kolom ke numerik jika memungkinkan (dengan NaN untuk nilai yang tidak dapat dikonversi)
data_numeric = data_reduced.apply(pd.to_numeric, errors='coerce')

# Identifikasi kolom yang tidak dapat dikonversi menjadi numerik, menandakan data kategorikal
non_numeric_cols = data_numeric.columns[data_numeric.isna().all()].tolist()

# Periksa data yang telah diubah untuk operasi numerik lebih lanjut
data_numeric_info = data_numeric.info()

non_numeric_cols, data_numeric_info
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np

# Menentukan jumlah klaster (K) yang optimal menggunakan Metode Elbow
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Membuat grafik Elbow untuk melihat K yang optimal
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Jumlah klaster (K)')
plt.ylabel('Inersia')
plt.title('Metode Elbow untuk K Optimal')
plt.show()

# Menerapkan KMeans dengan contoh K optimal (diasumsikan untuk pengujian lebih lanjut; akan disesuaikan setelah peninjauan pengguna)
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
cluster_labels = kmeans.fit_predict(data_scaled)

# Menerapkan PCA untuk visualisasi 2D
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Grafik sebar dengan pewarnaan berdasarkan kepadatan klaster
plt.figure(figsize=(8, 6))
scatter = plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cluster_labels, cmap="viridis", alpha=0.7)
plt.colorbar(scatter, label='Label Klaster')
plt.xlabel('Komponen PCA 1')
plt.ylabel('Komponen PCA 2')
plt.title(f'Klasterisasi K-Means (K={optimal_k}) - Proyeksi PCA')
plt.show()
```
